```python
import numpy as np
from mTAR import mTAR
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd

```

```python
phi1 = np.matrix([[0.5, 0.7], [0.3, 0.2]])
phi2 = np.matrix([[0.4, 0.6], [0.5, -0.5]])
sigma1 = np.matrix([[1, 0], [0, 1]])
sigma2 = np.matrix([[1, 0], [0, 1]])
c1 = np.array([0, 0])
c2 = np.array([0, 0])
delay = np.array([1, 1])
```

```python

y=mTAR(100,0,phi1,phi2,sigma1,sigma2,c1,c2,delay,ini=500)
mts = y.sim()
print(mts)

#print(mts['series'][:,0])
df = pd.DataFrame(mts['series'])
#print(df)
fig = go.Figure([go.Scatter(x= df.index, y=df[0])])
fig.add_trace(go.Scatter(x= df.index, y=df[1]))

fig.show()
```
